# Generate data.yaml for YOLO

Automatically generates the `data.yaml` configuration file required for YOLO model training.

The `data.yaml` file tells YOLO about:
- Path to training and validation images
- Number of classes (`nc`)
- Class names (`names`)

```python
import os
import yaml
```

## Configuration

```python
DATASET_PATH = "/content/drive/MyDrive/dataset_split"
CLASSES_PATH = os.path.join(DATASET_PATH, "classes.txt")
OUTPUT_PATH = os.path.join(DATASET_PATH, "data.yaml")

DEFAULT_CLASSES = [
    "Button", "Checkbox", "Dropdown", "Icon", "Image", "Input",
    "Link", "Menu", "Modal", "Progress_bar", "Radio_button",
    "Slider", "Tab", "Text", "Text_field", "Toolbar",
]
```

## Functions

```python
def load_classes(classes_path):
    """
    Reads class names from a classes.txt file.
    Each line corresponds to one class.
    Falls back to default classes if the file is not found.
    """
    if os.path.exists(classes_path):
        with open(classes_path, "r", encoding="utf-8") as f:
            classes = [line.strip() for line in f if line.strip()]
        print(f"classes.txt found with {len(classes)} classes.")
        return classes
    else:
        print(f"classes.txt not found at: {classes_path}")
        print(f"Using {len(DEFAULT_CLASSES)} default classes.")
        return DEFAULT_CLASSES


def generate_data_yaml(dataset_path, classes, output_path):
    """
    Generates the data.yaml file with the structure expected by YOLO.

    YAML structure:
        path: dataset root
        train: training images subfolder
        val: validation images subfolder
        test: test images subfolder
        nc: number of classes
        names: list of class names
    """
    yaml_data = {
        "path":  dataset_path,
        "train": "train/images",
        "val":   "val/images",
        "test":  "test/images",
        "nc":    len(classes),
        "names": classes,
    }

    with open(output_path, "w", encoding="utf-8") as f:
        yaml.dump(yaml_data, f, default_flow_style=False,
                  allow_unicode=True, sort_keys=False)

    print(f"\n data.yaml generated successfully!")
    print(f"Path: {output_path}")
    print(f"Classes: {len(classes)}")
    print(f"Names: {classes}")
```

## Run

```python
classes = load_classes(CLASSES_PATH)

for subfolder in ["train/images", "val/images"]:
    full_path = os.path.join(DATASET_PATH, subfolder)
    if os.path.isdir(full_path):
        num_imgs = len([f for f in os.listdir(full_path)
                       if f.lower().endswith((".png", ".jpg", ".jpeg"))])
        print(f"{subfolder}: {num_imgs} images found")
    else:
        print(f"Folder not found: {full_path}")

generate_data_yaml(DATASET_PATH, classes, OUTPUT_PATH)
```
